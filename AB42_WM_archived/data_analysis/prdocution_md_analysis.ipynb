## The following graphs illustrate CV space explored by the 48 replicas and the overall bias potential associated with the CV


<br>

```python
fig, ax = plt.subplots()
colvar_list = ["helix", "beta", "rgyr", "hydro", "salt", "dihcor", "bias"]
for i in range(48):
    data = pd.read_table(f"../system/simulations/replica{i}/COLVAR.{i}", delim_whitespace=True, skiprows=1, header=None)
    data_dict = {}
    time = data[0]
    data_dict["helix"] = data[1]
    data_dict["beta"] = data[2]
    data_dict["rgyr"] = data[3]
    data_dict["hydro"] = data[4]
    data_dict["salt"] = data[5]
    data_dict["dihcor"] = data[6]
    data_dict["bias"] = data[7]

    probability,xedges=np.histogram(data_dict["helix"], bins=100,range=[-1, 100],density=True,weights=None)
    xcenters= xedges[:-1] + np.diff(xedges)/2
    ax.plot(xcenters, probability, label=f"helix values - confirmation {i}")
    # ax.plot(helix_values, unknown_qty, label="uknown")
    ax.set_xlabel("Helix content")  # Add an x-label to the axes.
    ax.set_ylabel("Normalized Frequency")  # Add a y-label to the axes.
    ax.set_title("Production MD -- Alpha Helix Content Histogram of all 48 confirmations")
#     ax.legend()

data = pd.read_table("../system/simulations/replica0/GRID.helix.lessthan", delim_whitespace=True, skiprows=5, header=None)
values = data[0]
bias = data[1]

fig, ax = plt.subplots()
ax.plot(values, bias, label="bias")
ax.set_xlabel(f"helix")  # Add an x-label to the axes.
ax.set_ylabel("Bias")  # Add a y-label to the axes.
ax.set_title("Production MD (Confirmation 0)")
ax.legend()
```

```python
def plot_data(x, y, plot_label = "plot label", xlabel = "X axis Label", ylabel = "Y Label", title = "Graph Title"):
    fig, ax = plt.subplots()
    ax.plot(x, y, label=plot_label)
    ax.set_xlabel(xlabel)  # Add an x-label to the axes.
    ax.set_ylabel(ylabel)  # Add a y-label to the axes.
    ax.set_title(title)
    ax.legend()
```
