```python
data = pd.read_table("../system/simulations/confirmation0/npt_density.xvg", delim_whitespace=True, skiprows=24, header=None)
# dataA = dataA.drop(columns="Unnamed: 0")
x = data[0]
y = data[1]

sum = 0
avg = []
for i in range(len(y)):
    sum += y[i]
    avg.append(sum / (i+1))
```

```python
fig, ax = plt.subplots()
ax.plot(x, y)
ax.plot(x, avg, label="avg density")
ax.set_xlabel("time (ps)")  # Add an x-label to the axes.
ax.set_ylabel("Density (Kg/m^3)")  # Add a y-label to the axes.
ax.set_title("NPT Equilibration (Confirmation 0)")
ax.legend()
```
